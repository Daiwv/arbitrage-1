```python
import os

def file_length(file_name):
    with open(file_name) as file:
        for k, line in enumerate(file):
            None
        return k + 1
    
def get_currencies_from_file_list(file_list):
    result = set()
    for f in file_list:
        pair = f[0][:-4].split('-')
        x, y = pair[0], pair[1]
        if x not in result: result.add(x)
        if y not in result: result.add(y)
    return result

def get_files_w_length_gt(file_list, length):
    result = []
    for f in file_list:
        if f[1] > length: result.append(f)
    return result
```

```python
exchange = '../bfnx/'
files = []
for filename in os.listdir(exchange):
    if '-' in filename: # all my transaction pair files are in the form of base-quote.csv
        length = file_length(exchange + filename) - 1
        files.append((filename, length))
files
```

```python
currencies = get_currencies_from_file_list(files)
len(currencies), len(files)
```

```python
non_empty_files = get_files_w_length_gt(files, 0)
non_empty_currnecies = get_currencies_from_file_list(non_empty_files)
len(non_empty_currnecies), len(non_empty_files)
```

```python
files_gt_1k = get_files_w_length_gt(non_empty_files, 999)
currencies_gt_1k = get_currencies_from_file_list(files_gt_1k)
len(currencies_gt_1k), len(files_gt_1k)
```

```python
files_gt_10k = get_files_w_length_gt(files_gt_1k, 9999)
currencies_gt_10k = get_currencies_from_file_list(files_gt_10k)
len(currencies_gt_10k), len(files_gt_10k)
```
