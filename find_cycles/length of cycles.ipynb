```python
import numpy as np
```

```python
cycle_lengths = []
cycle_strings = []
with open('cycles_10k.txt', 'r') as cycles:
    for cycle in cycles:
        cycle_strings.append(cycle.strip())
        cycle_lengths.append(len(cycle.strip().split(',')))
```
